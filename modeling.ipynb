```python
# Lets remane some wierd columns name
books.rename(columns={"Book-Title":'title',
                      'Book-Author':'author',
                     "Year-Of-Publication":'year',
                     "Publisher":"publisher",
                     "Image-URL-L":"image_url"},inplace=True)
```

```python
books.head(1)
```

```python
# Lets remane some wierd columns name
users.rename(columns={"User-ID":'user_id',
                      'Location':'location',
                     "Age":'age'},inplace=True)
```

```python
users.head(1)
```

```python
# Lets remane some wierd columns name
ratings.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'rating'},inplace=True)
```

```python
ratings.head(1)
```

```python
# Lets store users who had at least rated more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y= x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
# Now join ratings with books

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.tail(1)
```

```python
# we don't need different size of images; we only need large size image to display
ratings_with_books = ratings_with_books.drop(['Image-URL-S', 'Image-URL-M'], axis=1)
```

```python
ratings_with_books.head(1)
```

```python
ratings_with_books.shape
```

```python
# select book titles and num of rating
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
# rename
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head(1)
```

```python
# combine all data
final_df = ratings_with_books.merge(number_rating, on='title')
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
# Lets take those books which got at least 30 rating of user

final_rating = final_df[final_df['num_of_rating'] >= 30]
```

```python
final_rating.shape
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

------------------

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
# convert NaN to 0
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

# Training Model

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors 
# it used to find the closest vectors 
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

# Testing

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
