# Collaborative Filtering Based

```python
books_with_ratings = ratings.merge(books, on='ISBN')
books_with_ratings.head(1)
```

```python
books_with_ratings.tail(1)
```

```python
books_with_ratings = books_with_ratings.rename(columns={"User-ID": "userID", 
                                   "Book-Rating": "Book_Rating",
                                   "Book-Title": "Book_Title",
                                   "Book-Author": "Book_Author",
                                   "Year-Of-Publication": "Year_Of_Publication",
                                   "Image-URL-S": "Image_URL_S",
                                   "Image-URL-M": "Image_URL_M",
                                   "Image-URL-L": "Image_URL_L"}
                                   )
```

```python
x = books_with_ratings.groupby("userID").count()['Book_Rating'] > 200
```

```python
# get the index
padhe_likhe_users = x[x].index
```

```python
# put the index into dataframe, and see the result
filtered_rating = books_with_ratings[books_with_ratings['userID'].isin(padhe_likhe_users)]
```

```python
filtered_rating.head(1)
```

```python
y = filtered_rating.groupby('Book_Title').count()['Book_Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book_Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book_Title', columns='userID', values='Book_Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

Users are columns; values are user's ratings. Each book can be represented as a vector, we can draw this vector in space. And then, we can find similar vectors based on computing cosine

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
# check the first book's similarity score with other books
similarity_scores[0]
# you notice the first value is 1, beacuse it's itself, so similarity score is 1
```

```python
similarity_scores[0].shape
```

```python
# find row number based on book title
np.where(pt.index=="You Belong To Me")[0]
```

```python
np.where(pt.index == "You Belong To Me")[0][0]
```

```python
# again, check first book's similarity score with other books
# this time, we sorted the list
sorted(list(enumerate(similarity_scores[0])), key=lambda x:x[1], reverse = True)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),
                           key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
# test first book, and see its most simliar books 
recommend('1984')
```

```python
books_with_ratings = ratings.merge(books, on='ISBN')
books_with_ratings.head(1)
```

```python
books_with_ratings.shape
```

```python
books_with_ratings = books_with_ratings.rename(columns={"User-ID": "userID", 
                                   "Book-Rating": "Book_Rating",
                                   "Book-Title": "Book_Title",
                                   "Book-Author": "Book_Author",
                                   "Year-Of-Publication": "Year_Of_Publication",
                                   "Image-URL-S": "Image_URL_S",
                                   "Image-URL-M": "Image_URL_M",
                                   "Image-URL-L": "Image_URL_L"}
                                   )
```

```python
books_with_ratings.head(1)
```

```python
books_with_ratings_df = books_with_ratings.groupby('Book_Title').count()['Book_Rating'].reset_index()
```

```python
books_with_ratings_df.tail(1)
```

```python
avg_books_with_ratings_df = books_with_ratings.groupby('Book_Title').mean()['Book_Rating'].reset_index()
avg_books_with_ratings_df.tail(1)
```

```python
popular_books_df = books_with_ratings_df.merge(avg_books_with_ratings_df, on="Book_Title")
```

```python
popular_books_df.head()
```

```python
popular_books_df = popular_books_df.rename(columns={"Book_Rating_x": "Book_Rating", 
                                                    "Book_Rating_y": "Avg_Rating"}
                                          )
```

```python
popular_books_df.head(1)
```

```python
most_popular_books_df = popular_books_df[popular_books_df['Book_Rating'] >= 250].sort_values('Avg_Rating', ascending=False)
```

```python
most_popular_books_df.head()
```

```python
most_popular_books_df.shape
```
